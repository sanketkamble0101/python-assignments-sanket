# Python OOP Assignment — Inheritance Examples

```python
# 1) Single Inheritance: Person -> Student
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, roll_no):
        super().__init__(name)
        self.roll_no = roll_no
        
        
s1 = Student("sanket",15)            # create object
print("name :",s1.name)
print("roll_no :",s1.roll_no)

# TODO:
# 1. Create object of Student.
# 2. Print name.
# 3. Print roll_no.
```

```python
# 2) Single Inheritance: Animal -> Dog
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")
        
d1 = Dog()      # create Dog Class Object
d1.sound()
d1.bark()

# TODO:
# 1. Create Dog object.
# 2. Call sound() from parent class.
# 3. Call bark() from child class.
```

```python
# 3) Constructor Inheritance: Vehicle -> Car
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model
        
        
c1 = Car('Toyota','Fortuner')  # <-- object
print("Brand :",c1.brand)
print("Model :",c1.model)


# TODO:
# 1. Create Car object.
# 2. Print brand.
# 3. Print model.
```

```python
# 4) Multilevel Inheritance: A -> B -> C
class A:
    def showA(self):
        print("This is A")

class B(A):
    def showB(self):
        print("This is B")

class C(B):
    def showC(self):
        print("This is C")
        

c1 = C()        # <-- object of C class

c1.showA()      # <-- call showA()
c1.showB()      # <-- call showB()
c1.showC()      # <-- call showC()

# TODO:
# 1. Create C object.
# 2. Call showA(), showB(), showC().
```

```python
# 5) Multilevel Inheritance with methods: LivingThing -> Animal -> Dog
class LivingThing:
    def living(self):
        print("I am alive")

class Animal(LivingThing):
    def eat(self):
        print("Eating…")

class Dog(Animal):
    def bark(self):
        print("Barking…")


d1 = Dog()

d1.living()
d1.eat()
d1.bark()
# TODO:
# 1. Create Dog object.
# 2. Call living(), eat(), bark().
```

```python
# 6) Multiple Inheritance: Father + Mother -> Child
class Father:
    def father_method(self):
        print("Father method")

class Mother:
    def mother_method(self):
        print("Mother method")

class Child(Father, Mother):
    def child_method(self):
        print("Child method")
        
c1 = Child()
c1.father_method()
c1.mother_method()
c1.child_method()

# TODO:
# 1. Create Child object.
# 2. Call father_method(), mother_method(), child_method().
```

```python
# 7) Constructor Inheritance: Product -> Laptop
class Product:
    def __init__(self, name):
        self.name = name

class Laptop(Product):
    def __init__(self, name, ram):
        super().__init__(name)
        self.ram = ram
        
l1 = Laptop("HP", "16GB")

print("Name :",l1.name)
print("Ram :",l1.ram)

# TODO:
# 1. Create Laptop object.
# 2. Print name.
# 3. Print ram.
```

```python
# 8) Multilevel with __init__: Country -> State -> City
class Country:
    def __init__(self, cname):
        self.cname = cname

class State(Country):
    def __init__(self, cname, sname):
        super().__init__(cname)
        self.sname = sname

class City(State):
    def __init__(self, cname, sname, cityname):
        super().__init__(cname, sname)
        self.cityname = cityname
        
        
city_obj = City("India","Maharashtra","Kolhapur")

print("Country :",city_obj.cname)
print("State :",city_obj.sname)
print("City :",city_obj.cityname)

# TODO:
# 1. Create City object.
# 2. Print cname, sname, cityname.
```

```python
# 9) Multiple Inheritance with methods: Teacher + Player -> Student
class Teacher:
    def teach(self):
        print("Teaching…")

class Player:
    def play(self):
        print("Playing…")

class Student(Teacher, Player):
    def study(self):
        print("Studying…")
        
stud1 = Student()
stud1.teach()
stud1.play()
stud1.study()

# TODO:
# 1. Create Student object.
# 2. Call teach(), play(), study().
```

```python
# 10) Constructor Inheritance: Shape -> Rectangle
class Shape:
    def __init__(self, color):
        self.color = color

class Rectangle(Shape):
    def __init__(self, color, length, width):
        super().__init__(color)
        self.length = length
        self.width = width
        
r1 = Rectangle("Blue",20,15)

print("Color :",r1.color)
print("Length :",r1.length)
print("Width :",r1.width)

# TODO:
# 1. Create Rectangle object.
# 2. Print color, length, width.
```
