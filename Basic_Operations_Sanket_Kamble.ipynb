```python
# index() --> Return the lowest index in String where substring sub is found.

str1 = "india"
str1.index('a')
```

```python
# isalnum() --> returns True if the string contains only alphabets and numbers, otherwise False.

print("Python123".isalnum())   # True
print("Python 123".isalnum())  # False  (space not allowed)
print("123".isalnum())         # True
print("Hello!".isalnum())      # False  (! not allowed)
```

```python
# isalpha() --> Returns True if the string has only alphabets (A–Z, a–z).

print("Python".isalpha())  # True
print("Python3".isalpha())  # False because of number
print("Hello World".isalpha())  # False because of space
```

```python
"""
isascii() --> returns True if all characters in the string are ASCII (0–127), otherwise False.

ASCII includes:
English letters (A–Z, a–z)
Digits (0–9)
Basic symbols (! @ # $ % …)
Spaces
Newlines

"""
print("Python123".isascii())      # True
print("Hello!".isascii())         # True
print("नमस्ते".isascii())         # False  (Hindi characters)
print(" ".isascii())              # True
```

```python
# isdecimal() --> returns True only if all characters in the string are decimal digits (0–9).

print("123".isdecimal())       # True
print("12.5".isdecimal())      # False  (dot not allowed)
print("½".isdecimal())         # False  (fraction symbol not allowed)
print("④".isdecimal())         # False  (circled numbers not allowed)
print("123a".isdecimal())      # False
```

```python
# isdigit() --> returns True if all characters in the string are digits, 
# including some special digit characters like superscripts and circled numbers.

print("123".isdigit())        # True
print("④".isdigit())          # True  # circled digit
print("²".isdigit())          # True  # superscript digit
print("12a".isdigit())        # False
print("12.5".isdigit())       # False  # dot not allowed
print("½".isdigit())          # False  # fraction not allowed
```

```python
"""
isidentifier() — What it checks?

A string is a valid identifier only if:
It starts with a letter (A-Z / a-z) or underscore _
After that, it can contain letters, digits, or underscores
No spaces
No special characters

"""
print("name".isidentifier())       # True
print("first_name".isidentifier()) # True
print("name1".isidentifier())      # True
print("_value".isidentifier())     # True

print("1name".isidentifier())      # False (cannot start with digit)
print("my-name".isidentifier())    # False (hyphen not allowed)
print("my name".isidentifier())    # False (space not allowed)
print("total$".isidentifier())     # False ($ not allowed)
print("for".isidentifier())        # True (BUT it is a keyword)
```

```python
# islower() --> Return True if the string is a lowercase string, False otherwise.

str1 = "india"
str1.islower()
```

```python
# isnumeric() --> Return True if the string is a numeric string, False otherwise.
# isnumeric() is the most powerful method to check numbers in a string.
# It accepts digits, superscripts, fractions, roman numerals, circled digits.
# Does not allow decimal points like "12.5"
# Does not allow letters

str1 = "india"
str1.isnumeric()
```

```python
# isprintable() --> is a Python string method that checks whether all characters in the string are printable.

print("Hello".isprintable())       # True (all characters visible)

print("Hello123".isprintable())    # True

print("Hello!".isprintable())      # True

print(" ".isprintable())           # True (space is printable)

print("Hello\nWorld".isprintable())  # False (\n is not printable)

print("Hello\tWorld".isprintable())  # False (\t is not printable)

print("".isprintable())            # True (empty string is printable)

print("₹100".isprintable())        # True (Indian Rupee symbol is printable)

```

Dictionary Operations

```python
dict = {"name":"sanket","marks":80}

dict
```

```python
# clear() --> removes all from dictionary

dict = {"name":"sanket","marks":80}
dict.clear()
dict
```

```python
# copy() --> makes a copy 

dict = {"name":"sanket","marks":80}

dict2 = dict.copy()

dict
```

```python
# fromkeys() --> Create a new dictionary with keys from iterable and values set to value.

key = ['name','location','grade']
dict.fromkeys(key)
```

```python
# get() --> Return the value for key if key is in the dictionary, else default.

dict = {"name":"sanket","marks":80}
dict.get('marks')
```

```python
# items() --> Return a set-like object providing a view on the dict's items.

dict.items()
```

```python
# keys() --> Return a set-like object providing a view on the dict's keys.
dict.keys()
```

```python
# pop() --> remove specified key and return the corresponding value.
# If the key is not found, return the default if given; otherwise, raise a KeyError.

dict = {"name":"sanket","marks":80}
dict.pop('name')
```

```python
# popitem() --> removes last item in a dictionary

dict = {"name":"sanket","marks":80}
dict.popitem()
```

```python
# If the key exists → returns its value
# if the key does NOT exist → adds the key with default_value

dict = {"name":"sanket","marks":80}
dict.setdefault("age",22)

dict
```

```python
# update() --> update values

dict = {"name":"sanket","marks":80}
dict.update({"name":"virat"})

dict
```

```python
# values() --> Return an object providing a view on the dict's values.


dict = {"name":"sanket","marks":80}
dict.values()
```

Set Operations

```python
set = {10, 3.14, "Sanket", True}

set
```

```python
set.add(8)

set
```

```python
set = {10, 3.14, "Sanket", True}
set.clear()
set
```

```python
set = {10, 3.14, "Sanket", True}
set2 = set.copy()

set2
```

```python
# difference() --> Return a new set with elements in the set that are not in the others.

set1 = {1,2,3,4}
set2 = {6,7,3,9}

result = set1.difference(set2)

result
```

```python
# difference_update() --> Update the set, removing elements found in others.

set1 = {1,2,3,4}
set2 = {6,7,3,9}
set1.difference_update(set2)

set1
```

```python
# discard() --> Remove an element from a set if it is a member.
# Unlike set.remove(), the discard() method does not raise an exception when an element is missing from the set.

set = {10, 3.14, "Sanket", True}
set.discard(3.14)

set
```

```python
# intersection() --> Return a new set with elements common to the set and all others.

set1 = {1, 2, 3, 4}
set2 = {6, 7, 3, 9}
result = set1.intersection(set2)

result
```

```python
# intersection_update() --> Update the set, keeping only elements found in it and all others.

set1 = {1, 2, 3, 4}
set2 = {6, 7, 3, 9}
set1.intersection_update(set2)

set1
```

```python
# isdisjoint() --> Return True if two sets have a null intersection.

set1 = {1, 2, 3, 4}
set2 = {6, 7, 3, 9}
result = set1.isdisjoint(set2)

result
```

```python
# issubset() --> To check whether all elements of one set exist in another set.

set1 = {1, 2, 3, 4}
set2 = {6, 7, 3, 9}
set1.issubset(set2)
```

```python
set1 = {1, 2, 3, 4}
set2 = {1, 2, 3, 4}
set1.issubset(set2)
```

```python
# set.issuperset() --> Report whether another set contains this set.

set1 = {1, 2, 3, 4}
set2 = {1, 2, 3}

set1.issuperset(set2)
```

```python

set = {1, 2, 3, 4}
set.pop()
```

```python
# set.remove() --> Remove an element from a set; it must be a member.If the element is not a member, raise a KeyError.

set = {1, 2, 3, 4}
set.remove(2)

set
```

```python
# set.symmetric_difference() --> Return a new set with elements in either the set or other but not both.

set1 = {1, 2, 3, 4}
set2 = {6, 7, 3, 9}
set1.symmetric_difference(set2)

```

```python
# set.union() --> Return a new set with elements from the set and all others.

set1 = {1, 2, 3, 4}
set2 = {6, 7, 3, 9}
set1.union(set2)
```

```python
# set.update() --> Update the set, adding elements from all others.

set1 = {1, 2, 3, 4}
set2 = {6, 7, 3, 9}

set1.update(set2)

set1
```
