### What does *args mean in Python? Write a function that accepts any number of values using *args and prints them.

```python
'''
In Python, *args is used in a function definition to allow the function to accept any number of positional arguments.
The values passed to *args are stored in a tuple, and you can loop through them or access them by index.
*args stands for “arbitrary arguments”, meaning the number of arguments is not fixed.

'''

def print_values(*args):
    for value in args:
        print(value)

# Calling the function with different number of arguments
print_values(10, 20, 30)
print_values("apple", "banana", "mango", 100)
```

### What does **kwargs mean? Write a function that prints all keyword arguments using **kwargs.

```python
'''
**kwargs stands for keyword arguments.
It allows a function to accept any number of keyword arguments (arguments in the form of key=value) without knowing them in advance.

kwargs is a dictionary that stores all the keyword arguments passed to the function.

** is used to unpack the dictionary.

'''

def print_kwargs(**kwargs) :
    for key,val in kwargs.items() :
        print(key,":",val)
        
print_kwargs(name="sanket",age=20,city="pune")
```

### Write an adapter function that takes a list of operations and executes add, sub, mul, div based on input. 

### Example: [ {"op": "add", "a": 10, "b": 20}, {"op": "mul", "a": 4, "b": 5} ]

```python
def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b

def div(a,b):
    return a/b

def select_option(option,ip):
    for choice in option:
        if choice['op'] == ip:
            a = choice['a']
            b = choice['b']
            if ip == 'add':
                return add(a,b)
            elif ip == 'sub':
                return sub(a,b)
            elif ip == 'mul':
                return mul(a,b)
            elif ip == 'div':
                return div(a,b)

lst = [ {"op": "add", "a": 10, "b": 20}, 
       {"op": "mul", "a": 4, "b": 5} ,
       {"op": "sub", "a": 14, "b": 5},
       {"op": "div", "a": 14, "b": 7}]

select_option(lst,'div')
```

### Write a function process_students(data) that prints total marks and grade for each student.
- Example: [ {"name":"A", "marks":[80,90,70]}, {"name":"B", "marks":[60,75,65]} ] 
- Expected Output: Name: A | Total: 240 | Grade: A Name: B | Total: 200 | Grade: B

```python
def process_students(data):
    for student in data:
        name = student["name"]
        marks= student["marks"]
        
        total= sum(marks)
        
        if total >= 240 :
            grade = 'A'
        elif total >= 180 :
            grade = 'B'
        elif total >= 150 :
            grade = 'C'
        else :
            grade = 'F'
        
        print(f"{name}|Total:{total}|Grade:{grade}")
        
 
        
stud_list = [ {"name":"A", "marks":[80,90,70]}, {"name":"B", "marks":[60,75,65]} ] 

process_students(stud_list)
```

### Create a function apply_discounts(products) that updates price after discount. 
- Example: [ {"name":"Laptop", "price":50000, "discount":10}, {"name":"Phone", "price":20000, "discount":5} ] 
- Expected Output: Laptop final: 45000 Phone final: 19000

```python
def apply_discount(products):
    for val in products:
        name = val['name']
        price = val['price']
        disc = val['discount']
        
        discount_price = price * disc // 100
        
        final_price = price - discount_price
        
        print(f"{name} final:{final_price}")


lst = [ {"name":"Laptop", "price":50000, "discount":10},{"name":"Phone","price":20000, "discount":5} ]

apply_discount(lst)
```

### Write a function filter_by_department(emp_list, dept) that returns employees belonging to a specific department.

```python
def filter_by_department(emp_list,dept):
    new_lst = []
    for details in emp_list:
        
        if details['department'] == dept :
            new_lst.append(details['name'])
            
    return new_lst      

lst = [
    {"id": 101, "name": "Amit", "age": 28, "department": "IT"},
    {"id": 102, "name": "Neha", "age": 24, "department": "HR"},
    {"id": 103, "name": "Raj", "age": 31, "department": "Finance"},
    {"id": 104, "name": "Simran", "age": 29, "department": "IT"},
    {"id": 105, "name": "Vijay", "age": 35, "department": "Admin"}
]

filter_by_department(lst,'IT')
```

### Write a function average_salary(emp_list) that returns the average salary.

```python
def average_salary(emp_list):
    count = 0
    sum = 0
    for details in emp_list:
        count += 1
        sum += details['salary']
        
    avg_salary = sum / count
    return f"average salary :{avg_salary}"

lst = [ {"id": 101, "name": "Amit", "age": 28, "department": "IT", "salary": 95000}, 
       {"id": 102, "name": "Neha", "age": 24, "department": "HR", "salary": 40000}, 
       {"id": 103, "name": "Raj", "age": 31, "department": "Finance", "salary": 65000}, 
       {"id": 104, "name": "Simran", "age": 29, "department": "IT", "salary": 60000}, 
       {"id": 105, "name": "Vijay", "age": 35, "department": "Admin", "salary": 35000} ]

average_salary(lst)
```

### Write out_of_stock(items) that returns all products with stock < 20.

```python
def out_of_stock(items):
    out_items = []
    for item in items:
        if item['stock'] < 20 :
            out_items.append(item['name'])
    return out_items

lst = [
    {"id": 1, "name": "Pen", "stock": 15},
    {"id": 2, "name": "Notebook", "stock": 50},
    {"id": 3, "name": "Pencil", "stock": 10},
    {"id": 4, "name": "Eraser", "stock": 5},
    {"id": 5, "name": "Marker", "stock": 30}
]

out_of_stock(lst)
```

### Write total_inventory_value(items) → sum(price * stock).

```python
def total_inventory_value(item):
    
    for i in item :
        total = 0
        item_value = i['price'] * i['stock']
        print(f"inventory value of {i['name']} = {item_value}")
        
        total = total + item_value  
        
    return f"Total Inventory Value = {total}"
    
lst = [
    {"name": "Pen", "price": 10, "stock": 50},
    {"name": "Notebook", "price": 40, "stock": 30},
    {"name": "Pencil", "price": 5, "stock": 100},
    {"name": "Marker", "price": 25, "stock": 20}
]

total_inventory_value(lst)
```

### Write search_product(items, name) → return matching product(s).

```python
def search_product(items,name):
    for details in items:
        if details['name'] == name:
            return details
            
    
    

lst = [
    {"name": "Pen", "price": 10, "stock": 50},
    {"name": "Notebook", "price": 40, "stock": 30},
    {"name": "Pencil", "price": 5, "stock": 100},
    {"name": "Marker", "price": 25, "stock": 20}
]

search_product(lst, "Pen")
```
