### 41. Given a dictionary of 10 items, delete all items whose value is less than 50 using a for loop

```python
products = {
    "mango": 80,
    "banana": 40,
    "apple": 120,
    "grapes": 150,
    "cherry": 20,
    "orange": 60,
    "watermelon": 180,
    "pineapple": 9,
    "kiwi": 130,
    "strawberry": 250
}

remove_keys = []

for fruit,price in products.items() :
    if price < 50 :
        remove_keys.append(fruit)
        
for key in remove_keys :
    del products[key]
    
products
```
