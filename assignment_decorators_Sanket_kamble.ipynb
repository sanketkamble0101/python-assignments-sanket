**Q22.** Explain (in a comment) why `convert_temperature()` should be an instance method if it depends on an object’s `scale` (C or F). Optionally, show a small example.

```python
# convert_temperature() should be an instance method because
# it depends on the object's attribute 'scale'. Each object can
# have a different scale (Celsius or Fahrenheit), so the method
# needs to know which object's scale it is converting.


class Temperature:
    def __init__(self, value, scale):
        self.value = value
        self.scale = scale.upper()  # 'C' or 'F'

    # Instance method because it depends on self.scale
    def convert_temperature(self):
        if self.scale == 'C':
            converted = (self.value * 9/5) + 32
            print(f"{self.value}°C is {converted}°F")
        elif self.scale == 'F':
            converted = (self.value - 32) * 5/9
            print(f"{self.value}°F is {converted}°C")
        else:
            print("Unknown scale")

# Test examples
t1 = Temperature(100, 'C')
t1.convert_temperature()   # 100°C is 212.0°F

t2 = Temperature(32, 'F')
t2.convert_temperature()   # 32°F is 0.0°C


```

**Q24.** Make a class `Date` with class method `from_string(cls, s)` to parse `"DD-MM-YYYY"` and return an object.

```python
class Date :

    def __init__(self,day,month,year):
        self.day = day
        self.month = month
        self.year = year
    
    @classmethod
    def from_string(cls,s):
        day,month,year =  map(int,s.split('-'))
        return cls(day,month,year)
    
    def display(self):
        print(f"Date: {self.day:02d}-{self.month:02d}-{self.year}")

        
            
        
        
# Test examples
d1 = Date.from_string("05-12-2025")
d1.display()  # Output: Date: 05-12-2025

d2 = Date.from_string("25-01-2023")
d2.display()  # Output: Date: 25-01-2023
```

**Q27.** Write a simple decorator that prints `"Function started"` before running the function. Test it on a small function.

```python
def decor(func):
    def innner (*args,**kwargs):
        print("Function Started")
        func(*args,**kwargs)
        return None
    return innner



# ----------------------------------------------------------------------------------

@decor
def intro():
    print("hello sanket")

intro()
```

**Q28.** Create a decorator that prints `"Function ended"` after running the function. Test it.

```python
def decor (func):
    def inner (*args,**kwargs):
        func(*args,**kwargs)
        print("function ended") 
        return None
    return inner

@decor
def intro():
    print("hello sanket")

intro()
```

**Q29.** Make a decorator that prints the **name** of the function being called (use `func.__name__`).

```python
def print_func_name(func):
    def wrapper(*args, **kwargs):
        print("Calling function:", func.__name__)
        return func(*args, **kwargs)
    return wrapper

@print_func_name
def greet():
    print("Hello, welcome!")

greet()
```

**Q30.** Combine two decorators: one prints `"Start"`, another prints `"End"` around a function. Check the order of messages.

```python
def start_decorator(func):
    def wrapper(*args, **kwargs):
        print("Start")
        return func(*args, **kwargs)
    return wrapper

def end_decorator(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        print("End")
        return result
    return wrapper

@start_decorator
@end_decorator
def show():
    print("Main Function")

show()


```
