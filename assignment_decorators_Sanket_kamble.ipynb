## B. Calling One Method Inside Another (6–10)


**Q6.** Create a class `Demo` with two methods: `method1()` prints `"Method 1"`, `method2()` calls `method1()` and then prints `"Method 2"`.

```python
class Demo:
    
    def method1(self):
        print("i am from method 1")
        
    def method2(self):
        self.method1()
        print("i am from method 2")
        
obj = Demo()
obj.method2()
```

**Q7.** Make a class `Math` with `square(n)` and `double_square(n)` that calls `square(n)` and doubles the result.

```python
class Math:
    def square(self,num):
        sq=num**2
        print(f"{num} square --> {sq}")
    
    def double_square(self,num):
        self.square(num)
        sqd = (num ** 2)*2
        print(f"double value is {sqd}")
    
obj = Math()
obj.double_square(4)

```

**Q8.** Create a class `Car` with `start_engine()` and `drive()` where `drive()` calls `start_engine()` first, then prints `"Driving..."`.

```python
class Car:
    def start_engine(self):
        print("engine started")
    
    def drive(self):
        self.start_engine()
        print("Driving...")
        
obj = Car()
obj.drive()

```

**Q12.** Make a class `Book` with `__init__(self, title, author)` and a method `show_info()` to display both.

**Q15.** Create a class `Account` with instance variables `name` and `balance`. Add method `show_balance()` to print balance.

```python
class Account:
    def __init__(self,name,balance):
        self.name = name               #instance variable
        self.balance = balance          # instance variable
        
    def show_balance(self):
        print("name is :",self.name)
        print("balance is :",self.balance)

obj = Account("sanket",50000)
obj.show_balance()
```
