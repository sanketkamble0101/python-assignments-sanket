### 40. Check if a number is Armstrong (3-digit).

```python
num = 153
original = num
count = len(str(num))
result = 0

while num > 0 :
    digit = num % 10
    result = result + digit ** count
    num = num // 10
    
if result == original :
    print("the number is armstrong")
else :
    print("number is not a armstrong number")
```
